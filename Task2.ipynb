```python
# Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        # Check minimum recharge amount
        if amount < 50:
            print("Recharge amount must be at least ₹50. Please try again.\n")

        # Check if amount matches valid plans
        elif amount not in valid_plans:
            print("Invalid recharge plan.")
            print("Available plans are: 199, 299, 399, 599\n")

        else:
            print(f"Recharge of ₹{amount} is successful ✅")
            break


# Function call
validate_recharge()
```

```python
# Inventory Reorder Alert System

def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert ⚠️ (Stock: {stock})")
        else:
            print(f"{product}: Stock OK ✅ (Stock: {stock})")


# Product stock dictionary
inventory = {
    "Mobile": 10,
    "Laptop": 25,
    "Headphones": 8,
    "Charger": 30
}

# Function call
check_inventory(inventory)
```

```python
# Student Result Processing System

def process_result(marks):
    total = 0

    # Calculate total marks using loop
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)

    # Check pass or fail
    if average >= 50:
        return f"Average Marks: {average:.2f} → Pass ✅"
    else:
        return f"Average Marks: {average:.2f} → Fail ❌"


# Sample student marks
student_marks = [60, 55, 70, 45, 50]

# Function call
result = process_result(student_marks)
print(result)
```

```python
# Cab Fare Estimator with Retry Option

def calculate_fare():
    base_fare = 50
    rate_per_km = 12

    while True:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        # Calculate base fare
        fare = base_fare + (distance * rate_per_km)

        # Apply peak hour surcharge
        if peak == "yes":
            fare += fare * 0.25

        print(f"Total Cab Fare: ₹{fare:.2f}")

        # Retry option
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break


# Function call
calculate_fare()
```

```python
# Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    total_days = len(attendance_list)
    attendance_percentage = (present_days / total_days) * 100

    # Check eligibility
    if attendance_percentage >= 75:
        return f"Attendance: {attendance_percentage:.2f}% → Eligible ✅"
    else:
        return f"Attendance: {attendance_percentage:.2f}% → Not Eligible ❌"


# Sample attendance data
attendance = ["P", "A", "P", "P", "P", "A", "P", "P"]

# Function call
result = check_attendance(attendance)
print(result)
```
