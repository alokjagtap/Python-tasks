```python
# Problem 1: Social Media – Post Engagement Analyzer
# Function to analyze post engagement
def analyze_post_engagement(likes_list):

    total_likes = 0

    # Loop to calculate total likes
    for like in likes_list:
        total_likes = total_likes + like

    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Display result
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Sample data (likes from different users)
likes = [200, 150, 300, 250, 150]

# Function call
analyze_post_engagement(likes)
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System
# Function to monitor medicine stock
def check_medicine_stock(stock):

    # Check stock level
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    # Display result
    print("Medicine Stock:", stock)
    print("Status:", status)


# Sample stock value
medicine_stock = 6

# Function call
check_medicine_stock(medicine_stock)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
# Function to check rainfall adequacy
def check_rainfall(rainfall_list, required_level):

    total_rainfall = 0

    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall = total_rainfall + rain

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)

    # Check rainfall status
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Display result
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)


# Sample seasonal rainfall data
rainfall_data = [70, 75, 68, 80, 67]
required_rainfall = 70

# Function call
check_rainfall(rainfall_data, required_rainfall)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
# Function to detect duplicate usernames
def detect_duplicate_accounts(usernames):

    # Convert list to set
    unique_usernames = set(usernames)

    # Check for duplicates
    if len(usernames) != len(unique_usernames):
        result = "Yes"
    else:
        result = "No"

    # Display result
    print("Duplicate Accounts Found:", result)


# Sample usernames list
user_list = ["aj123", "rahul99", "aj123", "neha_07", "rohit55"]

# Function call
detect_duplicate_accounts(user_list)
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker
# Function to check appointment eligibility
def check_appointment_eligibility(age):

    # Check eligibility
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    # Display result
    print("Patient Age:", age)
    print("Eligibility Status:", status)


# Sample patient age
patient_age = 21

# Function call
check_appointment_eligibility(patient_age)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
# Function to filter premium crop prices
def filter_premium_crops(price_list):

    premium_crops = []

    # Loop to filter premium prices
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)

    # Display result
    print("Premium Crops:", premium_crops)


# Sample crop prices
crop_prices = [1800, 2500, 1990, 3200, 1500]

# Function call
filter_premium_crops(crop_prices)
```
